#12. Outlier Detection and Handling:

Identify potential outliers in the 'Total Landed Weight' using the IQR (Interquartile Range) method.


```python
Q1 = df['Total Landed Weight'].quantile(0.25)
Q3 = df['Total Landed Weight'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Total Landed Weight'] < (Q1 - 1.5 * IQR)) | (df['Total Landed Weight'] > (Q3 + 1.5 * IQR))]
```

Remove the rows with outliers in the 'Landing Count' column.


```python
Q1=df['Landing Count'].quantile(0.25)
Q3=df['Landing Count'].quantile(0.75)
IQR=Q3-Q1
remove_row=df[(df['Landing Count'] >= (Q1 - 1.5 * IQR)) & (df['Landing Count'] <= (Q3 + 1.5 * IQR))]

```
